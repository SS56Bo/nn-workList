IMPORT LIBRARIES

```python
import numpy as np;
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

IMPORTING DATASETS

```python
dataSet = pd.read_csv('weightTrack.csv')
dataSet = pd.DataFrame(dataSet)
```

GET X-AXIS AND Y-AXIS

```python
X = dataSet.iloc[:,-3:]
x_axis = X.iloc[:,0]
y_axis1 = X.iloc[:,1]
y_axis2 = X.iloc[:,2]
```

DATA VISUALISATION

```python
fig, ax = plt.subplots(1, 2, figsize=(10, 5))
ax[0].scatter(x_axis, y_axis1, c='g', alpha=0.8)
ax[0].set_title('Days vs Weight (Day)')
ax[0].set_xlabel('Day')
ax[0].set_ylabel('Weights(in Kgs)')
ax[1].scatter(x_axis, y_axis2, c='r', alpha=0.8)
ax[1].set_title('Days vs Weight (Night)')
ax[1].set_xlabel('Day')
ax[1].set_ylabel('Weights(in Kgs)')
plt.tight_layout()
plt.show()
```
